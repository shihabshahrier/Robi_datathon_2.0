```python
melborn_data = melborn_data.dropna(axis=0)
```

```python
from sklearn.tree import DecisionTreeRegressor


filtered_data = melborn_data.dropna(axis = 0)
y = filtered_data.Price
melborn_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 
                        'YearBuilt', 'Lattitude', 'Longtitude']
x = filtered_data[melborn_features]

melborn_model = DecisionTreeRegressor()

melborn_model.fit(x, y)





```

```python
from sklearn.metrics import mean_absolute_error

predict_home_price = melborn_model.predict(x)
mean_absolute_error(y, predict_home_price)

```

```python
from sklearn.model_selection import train_test_split

train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 0)
melborn_model = DecisionTreeRegressor()
melborn_model.fit(train_x, train_y)
val_predict = melborn_model.predict(val_x)
mean_absolute_error(val_y, val_predict)


```

```python
def getMae(max_leaf_nodes, train_x, val_x, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state = 0)
    model.fit(train_x, train_y)
    pred = model.predict(val_x)
    mae = mean_absolute_error(val_y, pred)
    return mae
```

```python

for mln in [5, 50, 500, 4500]:
    mae = getMae(mln, train_x, val_x, train_y, val_y)
    print(f"max leaf nodes: {mln}   Mean Absolute error: {mae}")
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state = 1)
forest_model.fit(train_x, train_y)
pred = forest_model.predict(val_x)
print(mean_absolute_error(val_y, pred))
```
